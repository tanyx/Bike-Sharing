## Univariate and Multivariate Analysis

```python
# as we can see that season has numeric values.We will change this to string values so that we can create dummy variables
varlist = ['season']
bike[varlist] = bike[varlist].apply(lambda x:x.map({1:'spring', 2:'summer', 3:'fall', 4:'winter'}))
```

```python
# as we can see that weathersit has numeric values.We will change this to string values so that we can create dummy variables
varlist1 = ['weathersit']
bike[varlist1] = bike[varlist1].apply(lambda x:x.map({1:'Clear', 2:'Misty', 3:'Thunderstorm', 4:'Heavy Rain'}))
```

```python
# as we can see that mnth has numeric values.We will change this to string values so that we can create dummy variables
varlist2 = ['mnth']
bike[varlist2] = bike[varlist2].apply(lambda x:x.map({1:'January', 2:'February', 3:'March', 4:'April',5:'May', 6:'June', 7:'July', 8:'August',9:'September', 10:'October', 11:'November', 12:'December'}))
```

```python
# as we can see that weekday has numeric values.We will change this to string values so that we can create dummy variables
varlist3 = ['weekday']
bike[varlist3] = bike[varlist3].apply(lambda x:x.map({0:'Sunday', 1:'Monday', 2:'Tuesday', 3:'Wednesday', 4:'Thursday', 5:'Friday', 6:'Saturday'}))
```
